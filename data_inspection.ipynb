# JATO 数据结构分析

读取 JATO-2026.1.xlsx 文件，展示 Sheet 名称、数据形状、列名和样本数据。

```python
import pandas as pd
import openpyxl
from pathlib import Path

# 读取 Excel 文件
file_path = 'JATO-2026.1.xlsx'
excel_file = pd.ExcelFile(file_path)

# 所有 Sheet 名称
print('\n' + '='*60)
print('Excel Sheet 列表:')
print('='*60)
for i, sheet in enumerate(excel_file.sheet_names, 1):
    print(f'{i}. {sheet}')
```

```python
# 为每个 Sheet 读取数据并输出结构
for sheet_name in excel_file.sheet_names:
    print(f'\n{"="*60}')
    print(f'Sheet: {sheet_name}')
    print(f'{"="*60}')
    
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    print(f'\n数据度: {df.shape[0]} 行, {df.shape[1]} 列')
    print(f'\n列名 (类型):')
    for col in df.columns:
        print(f'  - {col}: {df[col].dtype}')
    
    print(f'\n第一衃数据:')
    print(df.head(3).to_string())
    
    if df.isnull().sum().sum() > 0:
        print(f'\n缺失值統計:')
        print(df.isnull().sum())
```

```python
# 快速汇总 - 每个 Sheet 的整体信息
print('\n' + '='*80)
print('数据结构汇总')
print('='*80)

for sheet_name in excel_file.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    print(f'\n[{sheet_name}]')
    print(f'  大小: {df.shape[0]:,} 行 × {df.shape[1]} 列')
    print(f'  列名: {list(df.columns)}')
    print(f'  数据类型: {dict(df.dtypes)}')
    null_count = df.isnull().sum().sum()
    if null_count > 0:
        print(f'  缺失值: {null_count} 个')
```

```python
# 详细查看第一个 Sheet: 透视表
sheet1_name = '\u900f视\u8868'
df_sheet1 = pd.read_excel(file_path, sheet_name=sheet1_name)

print(f'Sheet: {sheet1_name}')
print(f'\u5f62状: {df_sheet1.shape}')
print(f'\n列名 (总计 {len(df_sheet1.columns)} 列):')
print(df_sheet1.columns.tolist())

print(f'\n前 5 行数据:')
display(df_sheet1.head())
```

```python
# 详细查看第二个 Sheet: Data Export
sheet2_name = 'Data Export'
df_sheet2 = pd.read_excel(file_path, sheet_name=sheet2_name)

print(f'Sheet: {sheet2_name}')
print(f'\u5f62状: {df_sheet2.shape}')
print(f'\n列名 (总计 {len(df_sheet2.columns)} 列):')
print(df_sheet2.columns.tolist())

print(f'\n数据类型:')
print(df_sheet2.dtypes)

print(f'\n统计信息:')
print(df_sheet2.describe())

print(f'\n缺失值: {df_sheet2.isnull().sum().sum()}')

print(f'\n前 3 行:')
display(df_sheet2.head(3))
```

```python
# 简化汇总: 每个 Sheet 的关键信息
print('\n===== JATO Excel 数据结构简化 =====\n')

for sheet_name in excel_file.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    print(f'{sheet_name}')
    print(f'  算法: {df.shape[0]} 行 × {df.shape[1]} 列')
    print(f'  缺失值: {df.isnull().sum().sum()} 个')
    print(f'  数值列: {df.select_dtypes(include=["number"]).columns.tolist()}')
    print(f'  字符组列: {df.select_dtypes(include=["object"]).columns.tolist()}')
    print()
```
