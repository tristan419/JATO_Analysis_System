```python
import subprocess, sys
# 为当前 kernel 安装必要的包
packages = ['pandas', 'matplotlib', 'numpy', 'openpyxl', 'xlrd']
for pkg in packages:
    try:
        __import__(pkg)
        print(f'{pkg} 已安装')
    except ImportError:
        print(f'正在安装 {pkg}...')
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', pkg, '-q'])
        print(f'{pkg} 安装完成')
```
